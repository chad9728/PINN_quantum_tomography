# Comprehensive Scalability Analysis Figure Generation

This notebook generates a figure similar to FIG5 (fig:scalability) but based on the comprehensive scalability analysis table (tab:comprehensive_scalability).

The figure shows fidelity comparison across different dimensions (2-qubit to 10×10) for four methods:
- Traditional Neural Network (NN)
- Least Squares (LS)
- Maximum Likelihood Estimation (MLE)
- Physics-Informed Neural Network (PINN)


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 设置中文字体和样式
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")
sns.set_palette("husl")
```

```python
# Table3 (tab:comprehensive_scalability) 的数据
# 维度标签
dimensions = ['2-qubit', '3×3', '4×4', '5×5', '6×6', '7×7', '8×8', '9×9', '10×10']
# 维度数值（用于x轴）
dimension_values = [2, 3, 4, 5, 6, 7, 8, 9, 10]

# 四种方法的保真度数据
traditional_nn = [0.9764, 0.9207, 0.9894, 0.6731, 0.6594, 0.6273, 0.8427, 0.6035, 0.5401]
least_squares = [0.9346, 0.7626, 0.3433, 0.4684, 0.4260, 0.3719, 0.2141, 0.3010, 0.2781]
mle = [0.5281, 0.5605, 0.4041, 0.3793, 0.3284, 0.2706, 0.2219, 0.2147, 0.1879]
pinn = [0.9823, 0.9347, 0.9927, 0.6840, 0.6863, 0.6436, 0.8565, 0.6204, 0.5802]
```

```python
# 创建图形
fig, ax = plt.subplots(figsize=(14, 8))

# 绘制线条
line1 = ax.plot(dimension_values, traditional_nn, marker='o', markersize=8, linewidth=2.5, 
                label='Traditional NN', color='#2E86AB', linestyle='-', alpha=0.9)
line2 = ax.plot(dimension_values, least_squares, marker='s', markersize=8, linewidth=2.5, 
                label='Least Squares', color='#A23B72', linestyle='--', alpha=0.9)
line3 = ax.plot(dimension_values, mle, marker='^', markersize=8, linewidth=2.5, 
                label='MLE', color='#F18F01', linestyle='-.', alpha=0.9)
line4 = ax.plot(dimension_values, pinn, marker='D', markersize=10, linewidth=3, 
                label='PINN', color='#C73E1D', linestyle='-', alpha=0.9)

# 设置坐标轴
ax.set_xlabel('Dimension', fontsize=16, fontweight='bold')
ax.set_ylabel('Fidelity', fontsize=16, fontweight='bold')
ax.set_xticks(dimension_values)
ax.set_xticklabels(dimensions, fontsize=12, rotation=45, ha='right')
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_yticklabels([f'{y:.1f}' for y in np.arange(0, 1.1, 0.1)], fontsize=12)
ax.set_ylim([0, 1.05])
ax.set_xlim([1.5, 10.5])

# 添加网格
ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

# 添加图例
legend = ax.legend(loc='upper right', fontsize=13, frameon=True, 
                   fancybox=True, shadow=True, framealpha=0.9, ncol=1)
legend.get_frame().set_facecolor('white')
legend.get_frame().set_edgecolor('gray')

# 添加标题
ax.set_title('Comprehensive Scalability Analysis: Fidelity Comparison Across Dimensions', 
             fontsize=18, fontweight='bold', pad=25)


# 调整布局
plt.tight_layout()

# 保存图片
output_path = '../document/images/comprehensive_scalability_analysis.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight', facecolor='white')
print(f'图片已保存至: {output_path}')

# 显示图片
plt.show()
```

```python
# 计算统计信息
print("=" * 80)
print("Comprehensive Scalability Analysis Statistics")
print("=" * 80)
print(f"\nAverage Fidelity:")
print(f"  Traditional NN: {np.mean(traditional_nn):.4f}")
print(f"  Least Squares:  {np.mean(least_squares):.4f}")
print(f"  MLE:            {np.mean(mle):.4f}")
print(f"  PINN:           {np.mean(pinn):.4f}")

print(f"\nPINN Improvement over Traditional NN: {((np.mean(pinn) - np.mean(traditional_nn)) / np.mean(traditional_nn) * 100):.2f}%")
print(f"PINN Improvement over Least Squares: {((np.mean(pinn) - np.mean(least_squares)) / np.mean(least_squares) * 100):.2f}%")
print(f"PINN Improvement over MLE: {((np.mean(pinn) - np.mean(mle)) / np.mean(mle) * 100):.2f}%")

print(f"\nBest Performance at Each Dimension:")
for i, dim in enumerate(dimensions):
    methods = ['Traditional NN', 'Least Squares', 'MLE', 'PINN']
    values = [traditional_nn[i], least_squares[i], mle[i], pinn[i]]
    best_idx = np.argmax(values)
    print(f"  {dim:8s}: {methods[best_idx]:20s} ({values[best_idx]:.4f})")
```
